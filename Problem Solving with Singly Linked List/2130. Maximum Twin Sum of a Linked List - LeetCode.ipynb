### 2130. Maximum Twin Sum of a Linked List

```python
"""
In a linked list of size n, where n is even, the ith node (0-indexed) of the linked list is known as 
the twin of the (n-1-i)th node, if 0 <= i <= (n / 2) - 1.
For example, if n = 4, then node 0 is the twin of node 3, and node 1 is the twin of node 2. 
These are the only nodes with twins for n = 4.
The twin sum is defined as the sum of a node and its twin.
Given the head of a linked list with even length, return the maximum twin sum of the linked list.
Example 1:

Input: head = [5,4,2,1]
Output: 6
Explanation:
Nodes 0 and 1 are the twins of nodes 3 and 2, respectively. All have twin sum = 6.
There are no other nodes with twins in the linked list.
Thus, the maximum twin sum of the linked list is 6. 
Example 2:

Input: head = [4,2,2,3]
Output: 7
Explanation:
The nodes with twins present in this linked list are:
- Node 0 is the twin of node 3 having a twin sum of 4 + 3 = 7.
- Node 1 is the twin of node 2 having a twin sum of 2 + 2 = 4.
Thus, the maximum twin sum of the linked list is max(7, 4) = 7. 
Example 3:

Input: head = [1,100000]
Output: 100001
Explanation:
There is only one node with a twin in the linked list having twin sum of 1 + 100000 = 100001.
"""

# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

class Solution:
    def pairSum(self, head: Optional[ListNode]) -> int:
        curLength = self.__len__(head) # The total length of the Singly Linked List is returned
        threshold = (curLength // 2) - 1 # It is the limit, that should be taken into accout while calculating twins
        myDict = {} # The dictionary is created with indices as keys and nodes' values as dict_values
        count = 0
        current = head
        while current: # Let's start filling the dictionary with key-value pairs
            myDict[count] = current.val
            current = current.next
            count += 1
        twin_sums = [] # Let's create the container, which will collect twin sums
        index = 0
        while index <= threshold: # Here we are going to calculate the second twin node through the aforementioned 
            # formula
            first = myDict[index]
            second = myDict[curLength - 1 - index]
            curSum = first + second # The sum results in adding values of the first twin node and the second twin node together
            twin_sums.append(curSum) # All sums are added to the final list
            index += 1
        return max(twin_sums) # The maxElement is returned
    
    def __len__(self, head): # The function, which measures the total length of the Singly Linked List
        if head is None: # It is the case when the Singly Linked List is empty - we return 0
            return 0
        else:
            current = head  # Otherwise, all nodes are counted, while traversing the Singly Linked List
            count = 0
            while current:
                count += 1
                current = current.next 
            return count
```
