### 1670. Design Front Middle Back Queue

```python
"""
Design a queue that supports push and pop operations in the front, middle, and back.
Implement the FrontMiddleBack class:
FrontMiddleBack() Initializes the queue.
void pushFront(int val) Adds val to the front of the queue.
void pushMiddle(int val) Adds val to the middle of the queue.
void pushBack(int val) Adds val to the back of the queue.
int popFront() Removes the front element of the queue and returns it. If the queue is empty, return -1.
int popMiddle() Removes the middle element of the queue and returns it. If the queue is empty, return -1.
int popBack() Removes the back element of the queue and returns it. If the queue is empty, return -1.
Notice that when there are two middle position choices, the operation is performed on the frontmost middle position choice. For example:
Pushing 6 into the middle of [1, 2, 3, 4, 5] results in [1, 2, 6, 3, 4, 5].
Popping the middle from [1, 2, 3, 4, 5, 6] returns 3 and results in [1, 2, 4, 5, 6].
Example 1:

Input:
["FrontMiddleBackQueue", "pushFront", "pushBack", "pushMiddle", "pushMiddle", "popFront", "popMiddle", "popMiddle", "popBack", "popFront"]
[[], [1], [2], [3], [4], [], [], [], [], []]
Output:
[null, null, null, null, null, 1, 3, 4, 2, -1]

Explanation:
FrontMiddleBackQueue q = new FrontMiddleBackQueue();
q.pushFront(1);   // [1]
q.pushBack(2);    // [1, 2]
q.pushMiddle(3);  // [1, 3, 2]
q.pushMiddle(4);  // [1, 4, 3, 2]
q.popFront();     // return 1 -> [4, 3, 2]
q.popMiddle();    // return 3 -> [4, 2]
q.popMiddle();    // return 4 -> [2]
q.popBack();      // return 2 -> []
q.popFront();     // return -1 -> [] (The queue is empty)
"""

from collections import deque # With the help of 'collections.deque' we are going to use a double-ended queue

class FrontMiddleBackQueue:

    def __init__(self): # Constructor
        self.myQueue = deque()
        self.size = 0 # The initial size equals to 0

    def pushFront(self, val: int) -> None: # Method, which appends elements to the front of the queue
        self.myQueue.appendleft(val) 
        self.size += 1 # The size is increased by 1

    def pushMiddle(self, val: int) -> None: # Method, which adds elements to the middle of the queue
        middle = self.size // 2 # To find the middle address it is necessary to use the floor division
        self.myQueue.insert(middle, val)
        self.size += 1 # The size is increased by 1

    def pushBack(self, val: int) -> None: # Method, which appends elements to the end of the queue
        self.myQueue.append(val)
        self.size += 1 # The size is increased by 1

    def popFront(self) -> int: # Method, which extracts the front element
        if self.size == 0: # If there is nothing to extract - return -1
            return -1
        else: # In other case - extraction and size reduction by 1
            curElement = self.myQueue.popleft()
            self.size -= 1
            return curElement

    def popMiddle(self) -> int:
        if self.size == 0: # If there is nothing to extract - return -1
            return -1
        middle = self.size // 2
        if self.size % 2 == 0: # If the size is even-numbered - extraction of the element at index [middle index - 1]
            curElement = self.myQueue[middle - 1]
            del self.myQueue[middle - 1]
            self.size -= 1
            return curElement
        else: # As 'pop(index)' is not valid method with regard to 'collections.deque' we need to use 'del' to 
            # delete an element at the specific index
            curElement = self.myQueue[middle]
            del self.myQueue[middle]
            self.size -= 1 # Size reduction by 1
            return curElement

    def popBack(self) -> int:
        if self.size == 0: # If there is nothing to extract - return -1
            return -1
        else:
            # In other case, the standard list.pop() is used
            curElement = self.myQueue.pop()
            self.size -= 1 # Size reduction by 1
            return curElement
```
