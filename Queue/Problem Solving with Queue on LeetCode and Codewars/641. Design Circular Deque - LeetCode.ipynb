### 641. Design Circular Deque

```python
"""
Design your implementation of the circular double-ended queue (deque).
Implement the MyCircularDeque class:

MyCircularDeque(int k) Initializes the deque with a maximum size of k.
boolean insertFront() Adds an item at the front of Deque. Returns true if the operation is successful, or false otherwise.
boolean insertLast() Adds an item at the rear of Deque. Returns true if the operation is successful, or false otherwise.
boolean deleteFront() Deletes an item from the front of Deque. Returns true if the operation is successful, or false otherwise.
boolean deleteLast() Deletes an item from the rear of Deque. Returns true if the operation is successful, or false otherwise.
int getFront() Returns the front item from the Deque. Returns -1 if the deque is empty.
int getRear() Returns the last item from Deque. Returns -1 if the deque is empty.
boolean isEmpty() Returns true if the deque is empty, or false otherwise.
boolean isFull() Returns true if the deque is full, or false otherwise.

Example 1:

Input
["MyCircularDeque", "insertLast", "insertLast", "insertFront", "insertFront", "getRear", "isFull", "deleteLast", "insertFront", "getFront"]
[[3], [1], [2], [3], [4], [], [], [], [4], []]
Output
[null, true, true, true, false, 2, true, true, true, 4]

Explanation
MyCircularDeque myCircularDeque = new MyCircularDeque(3);
myCircularDeque.insertLast(1);  // return True
myCircularDeque.insertLast(2);  // return True
myCircularDeque.insertFront(3); // return True
myCircularDeque.insertFront(4); // return False, the queue is full.
myCircularDeque.getRear();      // return 2
myCircularDeque.isFull();       // return True
myCircularDeque.deleteLast();   // return True
myCircularDeque.insertFront(4); // return True
myCircularDeque.getFront();     // return 4
"""

from collections import deque # 'collections.deque' represents a double-ended queue

class MyCircularDeque:

    def __init__(self, k: int): # Constructor
        self.myQueue = deque()
        self.size = 0 # The initial number of elements within the queue structure
        self.maxVolume = k # The threshold
        

    def insertFront(self, value: int) -> bool:
        if self.isFull(): # If the queue already attained the envisaged threshold - we return False (
            # because we are unable to append additional elements)
            return False
        else:
            self.myQueue.appendleft(value) # The method, which appends elements to the front of the queue
            length = len(self.myQueue)
            if length > self.size:
                self.size += 1 # The number of elements within the queue structure is increased by 1
                return True
        
        
    def insertLast(self, value: int) -> bool:
        if self.isFull(): # If the queue already attained the envisaged threshold - we return False (
            # because we are unable to append additional elements)
            return False
        else:
            self.myQueue.append(value) # The method, which appends elements to the end of the queue
            length = len(self.myQueue)
            if length > self.size:
                self.size += 1 # The number of elements within the queue structure is increased by 1
                return True

    def deleteFront(self) -> bool:
        if self.isEmpty(): # If the queue is empty, there is nothing to delete, return False
            return False
        else:
            self.myQueue.popleft()  # Otherwise, deletion of the first element is implemented
            self.size -= 1 # The size is reduced by 1 respectively
            return True
        

    def deleteLast(self) -> bool:
        if self.isEmpty(): # If the queue is empty, there is nothing to delete, return False
            return False
        else:
            self.myQueue.pop() # Otherwise, deletion of the last element is implemented
            self.size -= 1 # The size is reduced by 1 respectively
            return True
            

    def getFront(self) -> int:
        if self.isEmpty(): # If the queue is empty, there is nothing to show/get - return -1
            return -1
        else:
            return self.myQueue[0] # Otherwise, we return the first element (index [0])

    def getRear(self) -> int:
        if self.isEmpty(): # If the queue is empty, there is nothing to show/get - return -1
            return -1
        else:
            return self.myQueue[-1] # Otherwise, we return the last element (index [-1])

    def isEmpty(self) -> bool: # If the number of elements within the queue equals to 0, the queue is empty, 
        # thus, return True
        if self.size == 0:
            return True
        else: # Otherwise, we return False
            return False
        

    def isFull(self) -> bool:
        # If the number of elements equals to the threshold - the queue is full - we return True
        if self.size == self.maxVolume:
            return True
        else:
            return False # Otherwise, we return False
```
