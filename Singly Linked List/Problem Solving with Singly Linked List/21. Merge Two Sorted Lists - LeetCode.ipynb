### 21. Merge Two Sorted Lists

```python
"""
You are given the heads of two sorted linked lists list1 and list2.
Merge the two lists in a one sorted list. 
The list should be made by splicing together the nodes of the first two lists.
Return the head of the merged linked list.
Example 1:
Input: list1 = [1,2,4], list2 = [1,3,4]
Output: [1,1,2,3,4,4]
Example 2:

Input: list1 = [], list2 = []
Output: []
Example 3:

Input: list1 = [], list2 = [0]
Output: [0]
"""

# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        if not list1 and not list2: # If both lists are empty, we return any of them
            return list1
        elif not list1: # If the first list is non-valid - we return the second list
            return list2
        elif not list2: # If the second list is non-valid - we return the first list
            return list1
        else:
            # After that, we are going to extract values of two Singly Linked Lists and add them to two new
            # initialized lists
            current1 = list1
            my_list1 = []
            while current1:
                my_list1.append(current1.val)
                current1 = current1.next
            current2 = list2
            my_list2 = []
            while current2:
                my_list2.append(current2.val)
                current2 = current2.next
            final = []
            i = 0
            j = 0
            # When two lists are filled with sorted values, the merge_sort algorithm is used (to merge two 
            # sorted arrays)
            while i < len(my_list1) and j < len(my_list2):
                if my_list1[i] > my_list2[j]:
                    final.append(my_list2[j])
                    j += 1
                else:
                    final.append(my_list1[i])
                    i += 1
            while i < len(my_list1):
                final.append(my_list1[i])
                i += 1
            while j < len(my_list2):
                final.append(my_list2[j])
                j += 1
            # When the final merged sorted array is formed, it is necessary to restore the Singly Linked List:
            list1 = ListNode(final[0])
            current = list1
            for i in range(1, len(final)):
                current.next = ListNode(final[i])
                current = current.next
            return list1
```
