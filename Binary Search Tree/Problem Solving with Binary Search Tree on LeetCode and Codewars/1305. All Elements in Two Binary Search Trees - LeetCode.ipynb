### 1305. All Elements in Two Binary Search Trees

```python
"""
Given two binary search trees root1 and root2, return a list containing all the integers from both trees 
sorted in ascending order.
Example 1:

Input: root1 = [2,1,4], root2 = [1,0,3]
Output: [0,1,1,2,3,4]
Example 2:

Input: root1 = [1,null,8], root2 = [8,1]
Output: [1,1,8,8]
"""

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right

class Solution:
    def getAllElements(self, root1: TreeNode, root2: TreeNode) -> List[int]:
        if root1 is None and root2 is None: # The case, when two Binary Search Trees are empty, the first one
            # or the second one
            return None
        # Depending on validity of Binary Search Tree we are going to implement the Inorder Traversal, which 
        # incorporates the sort of values in ascending order
        elif root1 and root2 is None:
            def first_BST(root1, final=[]): # All values in ascending order will be stored within the list container
                if root1.left:
                    first_BST(root1.left)
                curElement = root1.val
                final.append(curElement)
                if root1.right:
                    first_BST(root1.right)
                return final
            array1 = first_BST(root1)
            return sorted(array1)
        elif root1 is None and root2:
            def second_BST(root2, final=[]):
                if root2.left:
                    second_BST(root2.left)
                curElement = root2.val
                final.append(curElement)
                if root2.right:
                    second_BST(root2.right)
                return final
            array2 = second_BST(root2)
            return sorted(array2)
        else: # If two Binary Search Trees are not None data structures, we use two list containers to store
            # elements in ascending order
            def first_BST(root1, final=[]):
                if root1.left:
                    first_BST(root1.left)
                curElement = root1.val
                final.append(curElement)
                if root1.right:
                    first_BST(root1.right)
                return final
            array1 = first_BST(root1)
            
            def second_BST(root2, final=[]):
                if root2.left:
                    second_BST(root2.left)
                curElement = root2.val
                final.append(curElement)
                if root2.right:
                    second_BST(root2.right)
                return final
            array2 = second_BST(root2)
        # When two list containers are filled with elements, stored in ascending order, we use the Merge Sort
        # Algorithm, which will merge two sorted arrays into the one final sorted array
            i = 0
            j = 0
            result = []
            # The Merge Sort Algorithm to merge two sorted arrays into the one final sorted array is the following:
            while i < len(array1) and j < len(array2):
                if array1[i] > array2[j]:
                    result.append(array2[j])
                    j += 1
                else:
                    result.append(array1[i])
                    i += 1
            while i < len(array1):
                result.append(array1[i])
                i += 1
            while j < len(array2):
                result.append(array2[j])
                j += 1
            return result
```
